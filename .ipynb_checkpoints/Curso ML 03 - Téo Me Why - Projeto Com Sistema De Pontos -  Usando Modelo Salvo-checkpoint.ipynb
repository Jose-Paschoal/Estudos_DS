```python
## bibliotecas

import pandas as pd
```

## Lendo Modelo Salvo No Note Anterior

```python
model = pd.read_pickle('modelo_tree.pkl')

model
```

## Supondo Que Os Dados Do Note Anterior Sejam Novos Dados

```python
## importando os dados

df = pd.read_csv('dados_pontos.csv', sep=';')

df
```

## Predição Do Modelo

```python
X = df[model['features']]

predict_proba = model['model'].predict_proba(X)[:,1]  ## essa última marcação serve para filtrar apenas os que vão continuar

predict_proba
```

```python
df['prob_active'] = predict_proba

df[['Name', 'prob_active']]
```
